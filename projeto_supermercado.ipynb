```python
import pandas as pd
import plotly.express as px

# Carregar os dados
df = pd.read_csv('MODULO7_PROJETOFINAL_BASE_SUPERMERCADO - MODULO7_PROJETOFINAL_BASE_SUPERMERCADO (1).csv.csv')

# Limpeza e padronização
df.columns = df.columns.str.strip()
df['title'] = df['title'].str.strip().str.capitalize()
df['Marca'] = df['Marca'].str.strip().str.upper()

# Tratamento de Nans
df['Preco_Anterior'] = df.apply(
    lambda x: x['Preco_Normal'] if x['Preco_Anterior'] == 0 else x['Preco_Anterior'],
    axis=1
)
def calcular_percentual(row):
    if row['Preco_Anterior'] > 0:
        return (row['Desconto'] / row['Preco_Anterior']) * 100
    return 0
df['Valor_Desconto'] = df['Preco_Anterior'] - df['Preco_Desconto']
df.loc[df['Preco_Desconto'] == 0, 'Valor_Desconto'] = 0

# Criando as colunas necessárias
df['Percentual_Desconto'] = df.apply(calcular_percentual, axis=1)
df['Marca'] = df['Marca'].fillna('Sem Marca').str.upper()
df['Categoria'] = df['Categoria'].str.replace('-', ' ').str.title()

# Categoria x desconto
ranking_descontos = df.groupby('Categoria')['Percentual_Desconto'].mean().sort_values(ascending=False).reset_index()
print("Desconto Médio por Categoria:\n", ranking_descontos)

# Criar a coluna de % de Desconto Real
df['Porcentagem_Desconto'] = (df['Desconto'] / df['Preco_Anterior']) * 100
df['Porcentagem_Desconto'] = df['Porcentagem_Desconto'].fillna(0)

# Gráfico Interativo
fig = px.scatter(df,
                 x="Preco_Normal",
                 y="Porcentagem_Desconto",
                 color="Categoria",
                 hover_data=['title', 'Marca'],
                 title="Análise de Descontos por Categoria - Supermercado")

fig.show()

fig.update_layout(showlegend=False)
fig.show()
```
